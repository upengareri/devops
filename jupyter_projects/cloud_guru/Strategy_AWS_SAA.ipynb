## Further Notes

### AWS Training Exam Readiness (https://www.aws.training/Details/Curriculum?id=20685)

__Domains:__

__1. Designing Resilient Architecture__
    
    1.1 Choose resilient storage - as even in case of disaster your data is not lost
    
    1.2 Determine how to design decoupling mechanisms using AWS services - as it ensures if 1 tier or component fails, others are not impacted
    
    1.3 Determine how to design multi-tier architecture - as they are naturally decoupled and each scale independently
    
    1.4 Determine how to design HA and fault tolerant architecture
   
> Block storage like EBS allows for change in bytes whereas with object storage like S3 and Glacier we need to replace the entire object

Options for Resilient Storage:
- Instance Store: emphemeral and usually for caching and processing data for performance but not permanent
- EBS: decoupled from EC2
- EFS: one per VPC, Linux AMI only
- S3: standard S3 (where access is cheap but storage is expensive), S3-IA (where storage is cheap but access is expensive); 99.999999999 durability; encryption in transit by HTTP; encryption at rest by SSE-S3, SSE-KMS or SSE-C
- Glacier: encrypts data by default; best option is to have lifecycle policy to move data from S3 to glacier

Decoupling:
- SQS
- ELB

> RTO (Recovery Time Objective) is time to recover; RPO (Recovery Point Objective) is how much data is lost

Test Axioms on Resilient Architecture:
- Expect Single AZ will never be right answer
- Fault tolerant and HA are not same thing (HA says system is always up and can fail over on failure; Fault tolerant says system can conceal its failure from users and there is no loss of service)

__2. Design Performant Architectures__

2.1 Choose performant storage and databases

2.2 Apply caching

2.3 Design solutions for elasticity and scalibily

Options for Performant Storage:
- the first thing to choose for EBS storage is to go for SSD or HDD
- the other thing to do is to load all static contents to S3

S3 Payment Model:

_Pay what you use_
- GBs/month
- Transfer out of region
- API requests (GET, POST, PUT, COPY, LIST)

_Free_

- Transfer in to S3
- Transfer out from S3 to CloudFront or in same region

Options for Performant Database:
- RDS vs DynamoDB vs Redshift (when to choose what?)

Caching:
- CloudFront
- Elastic Cache: Memcached(simpler use case), Redis(complex use case)

> Remember with cache you don't want to put some data that is critical and needs update for the user e.g among shopping cart, product catalog and bank account balance, you should not store account balance in cache.

CloudFront:
- can be used both for static and dynamic contents
- dynamic content with TTL set to 0 but what is the benefit as we're not getting caching here. (benefit: so that Amazon's backbone network can be used for read and write)
- origins: S3 (for static), EC2, ELB, HTTP servers

Elasticity and Scalibility:
- ELB, ASG and CloudWatch
> Terminologies: vertical scaling (up and down), horizontal scaling (in and out)

__3. Secure Architecture__
3.1 IAM (users, groups, roles and policies)

3.2 VPC
- Organization: subnets
- Security: SG/NACL
- Network isolation: Internet gateways/virtual private gateways/NAT gateways
- Traffic direction: Routes

3.3 Securing Data Tier

_VPC_

Subnets: use subnets to define internet accessibility

_Public subnet:_ to support inbound/outbound traffic to public internet, include routing table entry to __internet gateway__

_Private subnet:_ do not have routing table entry to internet gateway; to support restricted outbound only public internet access, use a "jump-box" (NAT, proxy, bastion host)

SG vs NACL:
- sg applies to eni; nacl applies to subnet
- stateful; stateless
- explicit allow only; explicit allow or deny

Services to get traffic in and out of your VPC:
3. Internet gateway: enables connection to the internet
4. Virtual Private gateway: enables connection to the customer datacenter through vpn or direct connect
5. AWS Direct Connect: dedicated pipe b/w datacenter and AWS
1. VPC Peering: connect to other VPCs
2. NAT Gateways: allow outgoing internet traffic from private subnets

> Route tables are connected through subnets; subnets define connectivity

_Data Tier_

Data in Transit:
- in and out of AWS
- within AWS

Data at Rest:
- S3
- EBS

For protecting data in transit:
- use SSL over web
- VPN for IPsec (for data moving b/w AWS and datacenter)
- IPsec over AWS Direct Connect
- Import/Export service with snowball or snowmobile

> Data sent to AWS API are encrypted (HTTPS/SSL) by default

For protecting data at rest:
- S3 data is private by default
- Server side encryption: SSE-S3, SSE-KMS, SSE-C
- Client side encryption:
    - KMS Managed master key encryption (CSE-KMS)
    - Customer managed master key encryption (CSE-C)
    
> Client side encryption is mostly for compliance and regulation but server side encryption is easier and more performant

Where do you store your keys?

1. Key Management Service (KMS)
    - Customer software based key management
    - Use directly from application

2. CloudHSM
    - Hardware based key management
    - Use directly from application
    - FIPS 140-2 compliance

## Magic Numbers

- 100 S3 buckets per account
- 5 read replicas for RDS
- 128 compute nodes for RedShift
- There is a quota on the number of route tables per VPC as well as quota on the number of routes (rows) per route table
- 256KB size of messages in SQS
- 1 internet gateway per VPC
- 1 EFS per VPC
- Max VPCs per region = 5
- 1 NAT Gateway in 1 AZ
- 1 subnet in 1 NACL but 1 NACL can have many subnets
- many to many relationship b/w ec2 and sg
- max 7 running instances per AZ for spread PG
- messages in SQS can be b/w 1 min to 14 days; default retention period = 4 days
- max visibility timeout for a msg in sqs in 12 hrs
- while SQS retention period is 14days; SWF retention period is 1 year
- Retention period of media stream data in kinesis streams is b/w 24 hrs to 7 days
- Route53 default limit to number of domains = 50
- instance RAM < 150 GB for hibernation
- Instances can't be hibernated for > 60 days

## Checkmarks
- "Delete on termination" is checked for root EBS volumes attached to EC2 while for additional EBS (non-root volumes) it is not checked.
- Disable "Source/Destination" check while creating NAT Instance
- User created VPC subnet has "Auto assign public IP" unchecked
